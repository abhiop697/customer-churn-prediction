```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("../data/customer_churn.csv")

# Show dataset info
df.info()

# Check for missing values
print(df.isnull().sum())

# Churn distribution
sns.countplot(x=df['Churn'])
plt.title("Churn Distribution")
plt.show()
```
